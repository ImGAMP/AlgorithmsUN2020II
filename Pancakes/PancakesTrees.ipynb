```
import pydot
import time, random
from collections import defaultdict
from IPython.display import Image, display
import networkx as nx
import matplotlib.pyplot as plt
```

```
def lstr(p):
  return ''.join([str(elem) for elem in p])
```

```
def ModifiedBFS(s):
  tik = time.time()
  #Grafo como diccionario
  graph = defaultdict()
  visited = defaultdict()
  G = pydot.Dot(graph_type = "digraph")

  #Crear cola
  queue = []

  #Nodo naciente
  queue.append(s)
  visited[lstr(s)] = True

  while queue:

    #Sacar el vertice de la cola
    s = queue.pop(0)
    graph[lstr(s)]=[]
    node = pydot.Node(lstr(s))
    G.add_node(node)

    #Get all adjacent vertices of the 
    #dequeued vertex s. If a adjacent 
    #has not been visited, then mark it 
    #visited and enqueue it

    if (s != sorted(s)):
      for i in range(len(s)-1): 
        d = s[i+1::-1] + s[i+2:]
        if (d != s) and not(lstr(d) in visited.keys()):

          if (d != sorted(d)):
            node = pydot.Node(lstr(d))
          else:
            node = pydot.Node(lstr(d), style="filled", fillcolor="green") #Primer ordenado será verde

          G.add_node(node)
          edge = pydot.Edge(lstr(s),lstr(d))
          G.add_edge(edge)
          graph[lstr(s)].append(lstr(d))
          queue.append(d)
          visited[lstr(d)] = True
  tok = time.time()
  #Mostrar grafico pydot G
  im = Image(G.create_png())
  display(im) 
  return (tok-tik)* 1000
```

```
def case(f):
  a=[]
  b=[]
  for h in range(f):
    b.append(h)
  for m in range(len(b)):
    x = random.randrange(len(b))
    a.append(b[x])
    b.remove(b[x])
  return a
```

```
for i in range(1,7):
  p = case(i)
  tme = ModifiedBFS(p)
  print("Pn time for graph {0} is : {1}".format( i , tme))

```

```
def ModifiedBFSPreSols(s):
  tik = time.time()
  #Grafo como diccionario
  graph = defaultdict()
  visited = defaultdict()
  G = pydot.Dot(graph_type = "digraph")

  #Crear cola
  queue = []

  #Nodo naciente
  queue.append(s)
  visited[lstr(s)] = True

  while queue:

    #Sacar el vertice de la cola
    s = queue.pop(0)
    graph[lstr(s)]=[]

    #Crear bandera
    
    flag = False
    for i in range(len(s)-1):
      d = s[i+1::-1]+s[i+2:]
      if (d==sorted(d)):
        flag=True
    if flag:
      node = pydot.Node(lstr(s), style="filled", fillcolor= "yellow")
    else: 
      node = pydot.Node(lstr(s))

    G.add_node(node)

    #Get all adjacent vertices of the 
    #dequeued vertex s. If a adjacent 
    #has not been visited, then mark it 
    #visited and enqueue it

    if (s != sorted(s)):
      for i in range(len(s)-1): 
        d = s[i+1::-1] + s[i+2:]
        if (d != s) and not(lstr(d) in visited.keys()):

          if (d != sorted(d)):
            node = pydot.Node(lstr(d))
          else:
            node = pydot.Node(lstr(d), style="filled", fillcolor="green") #Primer ordenado será verde

          G.add_node(node)
          edge = pydot.Edge(lstr(s),lstr(d))
          G.add_edge(edge)
          graph[lstr(s)].append(lstr(d))
          queue.append(d)
          visited[lstr(d)] = True

  #Mostrar grafico pydot G
  tok = time.time()
  im = Image(G.create_png())
  display(im) 
  return (tok-tik)* 1000
```

```
def ModifiedBFSPreSolsLevel(s,L):
  tik = time.time()
  #Grafo como diccionario
  graph = defaultdict()
  visited = defaultdict()
  G = pydot.Dot(graph_type = "digraph")

  #Crear cola
  queue = []

  #Nodo naciente
  queue.append([s,0])
  visited[lstr(s)] = True

  while queue:

    #Sacar el vertice de la cola
    e = queue.pop(0)
    s = e[0]
    l = e [1]
    graph[lstr(s)]=[]

    #Crear bandera
    
    flag = False
    for i in range(len(s)-1):
      d = s[i+1::-1]+s[i+2:]
      if (d==sorted(d)):
        flag=True
    if flag:
      node = pydot.Node(lstr(s), style="filled", fillcolor= "yellow")
    else: 
      node = pydot.Node(lstr(s))

    G.add_node(node)

    #Get all adjacent vertices of the 
    #dequeued vertex s. If a adjacent 
    #has not been visited, then mark it 
    #visited and enqueue it

    if (s != sorted(s)) and  (l<L) :
      for i in range(len(s)-1): 
        d = s[i+1::-1] + s[i+2:]
        if (d != s) and not(lstr(d) in visited.keys()):

          if (d != sorted(d)):
            node = pydot.Node(lstr(d))
          else:
            node = pydot.Node(lstr(d), style="filled", fillcolor="green") #Primer ordenado será verde

          G.add_node(node)
          edge = pydot.Edge(lstr(s),lstr(d))
          G.add_edge(edge)
          graph[lstr(s)].append(lstr(d))
          queue.append([d,l+1])
          visited[lstr(d)] = True
  tok = time.time()
  #Mostrar grafico pydot G
  im = Image(G.create_png())
  display(im) 
  return (tok - tik)*100
```

```
for i in range(1,6):
  p = case(i)
  for j in range(i):
    tme = ModifiedBFSPreSolsLevel(p,j)
    print("Pn time for graph {0} with level {1} is : {2}".format( i , j , tme))
```

```
def NetModifiedBFS(s):
  #Grafo como diccionario
  graph = defaultdict()
  visited = defaultdict()
  G = nx.Graph() 

  #Crear cola
  queue = []

  #Nodo naciente
  queue.append(s)
  visited[lstr(s)] = True

  while queue:

    #Sacar el vertice de la cola
    s = queue.pop(0)
    graph[lstr(s)]=[]
    node = lstr(s)
    G.add_node(node)

    #Get all adjacent vertices of the 
    #dequeued vertex s. If a adjacent 
    #has not been visited, then mark it 
    #visited and enqueue it

    if (s != sorted(s)):
      for i in range(len(s)-1): 
        d = s[i+1::-1] + s[i+2:]
        if (d != s) and not(lstr(d) in visited.keys()):
          G.add_node(lstr(d))
          G.add_edge(lstr(s),lstr(d))
          graph[lstr(s)].append(lstr(d))
          queue.append(d)
          visited[lstr(d)] = True
          

  label = nx.nodes(G)
  nx.draw_shell(G, with_labels=True)
  plt.show()
  return G
```
